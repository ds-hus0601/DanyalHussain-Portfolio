##Statistical Anomaly Detection using IQR

To assess the severity of outlier overlap across features, the number and percentage of rows where 1+, 2+, 3+ and 4+ features were simultaneously flagged as outliers and cakculated below. It can be seen that IQR is effective at identifying feature-level anomalies but cannot capture inter-feature dependencies

```python
def detect_iqr_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

for feature in features:
    df[f"{feature}_outlier"] = detect_iqr_outliers(df[feature]).astype(int)

outlier_cols = [f"{feature}_outlier" for feature in features]
df["Anomaly_IQR"] = (df[outlier_cols].sum(axis=1) >= 2).astype(int)

anomaly_count = df["Anomaly_IQR"].sum()
print(f"IQR Anomalies in 2 or more features:", anomaly_count, f"({round(anomaly_count / len(df) * 100, 2)}%)")

for t in [1, 3, 4]:
  count = (df[outlier_cols].sum(axis=1) >= t).sum()
  percentage = (count / len(df)) * 100
  print(f" IQR Anomalies in {t} or more features: {count} samples ({percentage:.2f}%)")
```

### Reflections and Observations on IQR Method

The IQR method was useful in identifying individual feature-level anomalies, particularly those that fall well outside the expected value ranges.

While it is excellent at capturing feature-specific outliers, for example, unusally high RPM or Oil Temperature, it lacks the ability to consider interactions between multiple features. This is because it is fundamentally a univariate technique, treating each feature independently. As a result, it cannot detect contextual anomalies where normal values across multiple features might, in combination, signal abnormal engine behaviour.

To bridge this gap, evaluation on how many features simultaneously flagged a row as an outlier was considered. Thresholding the outlier count at 1+, 2+, 3+, and 4+ features allowed for a nuanced interpretation. For example, a sample flagged in 3 or more features is more likely to reflect a genuine anomaly of concern than a sample flagged in only one feature.

Additionally, by using a fixed threshold (1.5. * IQR), it may miss contextually significant variations that appear normal in isolation but are abnormal holistically. This would further highlight the need for a multivariate approach, like Isolation Forest.

In summary, while the IQR method successfully flagged 422 samples at a 2+ feature threshold (2.16% of the data), its simplistic view of the features risks both false positives and false negatives in operational environments where multivariate dependencies drive anomalies. Therefore, while useful for preliminary screening, it would not be sufficient as a standalone anomaly detection method in real-world deployments without more flexible, multivariate models.

The obervations here strongly justify the transition towards multivariate detection methods, such as One-Class SVM and Isolation Forest, which can capture nonlinear relationshops and interaction effect across multiple features simultaneously.

```python
features = [
    'Engine rpm',
    'Lub oil pressure',
    'Fuel pressure',
    'Coolant pressure',
    'lub oil temp',
    'Coolant temp'
]

df_fe = df[features].copy()
df_fe["temp_diff"] = df_fe["Coolant temp"] - df_fe["lub oil temp"]
df_fe["rpm_to_fuel_pressure"] = df_fe["Engine rpm"] / (df_fe["Fuel pressure"] + 1e-3)
df_fe["total_pressure"] = (
    df_fe["Lub oil pressure"] +
    df_fe["Fuel pressure"] +
    df_fe["Coolant pressure"]
)
df_fe["coolant_to_oil_pressure"] = df_fe["Coolant pressure"] / (df_fe["Lub oil pressure"] + 1e-3)
df_fe["avg_temp"] = (df_fe["lub oil temp"] + df_fe["Coolant temp"]) / 2
df_fe.head()
```

## Machine Learning Anomaly Detection - Without feeding engineered data into the models

```python
scaler = StandardScaler()
X = df[["Engine rpm", "Lub oil pressure", "Fuel pressure", "Coolant pressure", "lub oil temp", "Coolant temp"]]
X_scaled = scaler.fit_transform(X)
X_unscaled = X


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
###Testing different nu levels to tune One-Class SVM
for nu in [0.01, 0.02, 0.05]:
    svm = OneClassSVM(kernel="rbf", gamma=0.01, nu=nu)
    preds = svm.fit_predict(X_scaled)
    anomaly_rate = (preds == -1).mean()
    anomalies_count_svm = (preds == -1).sum()
    print(f"nu={nu}, Anomaly Rate: {anomaly_rate:.2%}, Number of Anomalies: {anomalies_count_svm}")
```

```python
###Final One-Class SVM Model (nu = 0.02)
svm = OneClassSVM(kernel="rbf", gamma=0.01, nu=0.02)
svm.fit(X_scaled)
df["Anomaly_SVM"] = (svm.predict(X_scaled) == -1).astype(int)

num_anomalies_svm = df["Anomaly_SVM"].sum()

normal_indices = df[df["Anomaly_SVM"] == 0].index
anomaly_indices = df[df["Anomaly_SVM"] == 1].index

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[normal_indices, 0], X_pca[normal_indices, 1], c='steelblue', label='Normal', alpha=0.6, edgecolor='black')
plt.scatter(X_pca[anomaly_indices, 0], X_pca[anomaly_indices, 1], c='darkred', label='Anomaly', alpha=0.7, s=18, edgecolor='black')
plt.legend(loc='lower right')
plt.title("One-Class SVM without Feature Engineering (PCA 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
print("Number of anomalies:", num_anomalies_svm)
```

```python
###Testing different contamination values
for contamination in [0.01, 0.02, 0.05]:
    iso = IsolationForest(contamination=contamination, random_state=42)
    preds = iso.fit_predict(X_unscaled)
    anomaly_rate = (preds == -1).mean()
    anomalies_count_if = (preds == -1).sum()
    print(f"contamination={contamination}, Anomaly Rate: {anomaly_rate:.2%}, Number of Anomalies: {anomalies_count_if}")
```

```python
#Final Isolation Forest Model (Contamination = 0.02)
iso = IsolationForest(contamination=0.02, random_state=42)
df["Anomaly_IF"] = (iso.fit_predict(X_unscaled) == -1).astype(int)

num_anomalies_if = df["Anomaly_IF"].sum()

normal_indices_if = df[df["Anomaly_IF"] == 0].index
anomaly_indices_if = df[df["Anomaly_IF"] == 1].index

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[normal_indices_if, 0], X_pca[normal_indices_if, 1], c='steelblue', label='Normal', alpha=0.6, edgecolor='black')
plt.scatter(X_pca[anomaly_indices_if, 0], X_pca[anomaly_indices_if, 1], c='darkred', label='Anomaly', alpha=0.7, s=18, edgecolor='black')
plt.legend(loc='lower right')
plt.title("Isolation Forest without Feature Engineering (PCA 2D)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.show()
print("Number of anomalies:", num_anomalies_if)
```
