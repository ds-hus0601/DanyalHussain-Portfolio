```python

import pandas as pd
import numpy as np
from statsmodels.stats.proportion import proportions_ztest
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv('https://github.com/FourthRevGitHub/CAM_DS_Public-Data-Sets/raw/main/Course%201/Week%206/CUSTOMERS_CLEAN.zip')

def clean_currency(val):
    if isinstance(val, str):
        val = val.strip().replace(',', '').replace('$', '').replace('(', '-').replace(')', '')
    return float(val)

currency_cols = ['Total Revenue', 'Unit Cost', 'Profit']
for col in currency_cols:
    df[col] = df[col].apply(clean_currency)

df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='%d%b%Y', errors='coerce')
df['Delivery_Date'] = pd.to_datetime(df['Delivery_Date'], format='%d%b%Y', errors='coerce')
df['Customer_BirthDate'] = pd.to_datetime(df['Customer_BirthDate'], format='%d%b%Y', errors='coerce')

reference_date = df['Order_Date'].max()
df['Customer_Age'] = (reference_date - df['Customer_BirthDate']).dt.days // 365
```

```python
df.head()
```

```python
missing_values = df.isnull().sum()
duplicates = df.duplicated().sum()
missing_percent = (missing_values / len(df)) * 100
print("Missing Values per Column:\n", missing_values[missing_values > 0] if missing_values.sum() > 0 else "No missing values found.")
print("\nMissing Values Percentage:\n", missing_percent[missing_percent > 0].round(2))
print("\nDuplicate Rows:", f"{duplicates} duplicates found." if duplicates > 0 else "No duplicate rows found.")
```

```python
features_to_test = ['City', 'Postal_Code', 'State_Province']
total_rows = len(df)
null_hypothesis = "No significant difference in the proportion of missing values"
alternate_hypothesis = "Significant difference in the proportion of missing values"

for feature in features_to_test:
    missing_count = df[feature].isnull().sum()
    count = np.array([missing_count])
    number_of_observations = np.array([total_rows])
    stat, p_value = proportions_ztest(count, number_of_observations, value=0.05, alternative='larger')

    if p_value < 0.05:
        print(f"\n{feature} - {alternate_hypothesis}")
    else:
        print(f"\n{feature} - {null_hypothesis}")

```

Missing values were detecting in City, Postal Code, and State Province. To assess their impact, a proportion Z-test was conducted to determine whether the proportion of missing values in each column was statistically greater than the acceptable threshold of 5%. The test confirmed that only State_Province had a signficantly higher proportion of missing data. However, as the State_Province feature is not directly relevant to the clustering variables, no imputation/dropping was needed. The remaining missing values were minimal as they showed no statistically significant deviation, and hence do not affect the clustering process. Therefore, the dataset was used as-is for segmentation.

Note: The Z-test was conducted with a 5% expected missing rate threshold using a one-sided hypothesis

```python
raw_numeric_cols = ['Quantity', 'Total Revenue', 'Unit Cost', 'Profit', 'Days to Delivery']
plt.figure(figsize=(15, 10))
for i, feature in enumerate(raw_numeric_cols):
    plt.subplot(3, 2, i + 1)
    sns.boxplot(data=df, x=feature, color='skyblue')
    plt.title(f'{feature} Boxplot')
    plt.tight_layout()
plt.suptitle("Initial EDA – Outlier Detection in Raw Numeric Features", fontsize=14, y=1.02)
plt.show()

def iqr_outlier_summary(df, columns):
    summary = []
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower) | (df[col] > upper)][col]
        count = outliers.count()
        percent = round((count / len(df)) * 100, 2)
        summary.append((col, count, f"{percent}%"))
    return pd.DataFrame(summary, columns=['Feature', 'Outlier Count', 'Outlier %'])

iqr_results = iqr_outlier_summary(df, raw_numeric_cols)
display(iqr_results)
```

To better understand the distribution and potential outliers within the dataset, an initial EDA was conducted on the key numeric features, namely: Quantity, Total Revenue, Unit Cost, Profit, and Days to Delivery.

The Boxplots revealed moderate skewness in features like Profit, Total Revenue, and Days to Delivery. These plots visually flagged the presence of outliers, likely driven by transactions including bulk orders or international shipping delais.

Whilst Days to Delivery shows the highest outlier rate, this is likely attributable, as discussed early, to logistics or rarer cases like pre-orders or backlogs/bulk orders.

Features like Profit and Revenue show dispersion consistent with typical marketing campaigns.

Following the Intial EDA, the outliers were retained for clustering. These values are meaningful and may improve the segmentation process by capturing all customer behaviour

```python

customer_df = df.groupby('Customer ID').agg(
    Frequency=('Order ID', 'nunique'),
    Recency=('Order_Date', lambda x: (reference_date - x.max()).days),
    CLV=('Profit', 'sum'),
    Avg_Unit_Cost=('Unit Cost', 'mean'),
    Customer_Age=('Customer_Age', 'mean')
).reset_index()
```

```python

features = ['Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Customer_Age']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(customer_df[features])
```

```python

inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(range(2, 11), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    silhouette_scores.append(score)
    print(f"k = {k:<2} -> Silhouette Score: {score:.4f}")



plt.figure()
plt.plot(range(2, 11), silhouette_scores, marker='s')
plt.title("Silhouette Score")
plt.xlabel("Number of Clusters")
plt.ylabel("Score")
plt.grid(True)
plt.show()

```

Based on the Elbow method and silhouette score analysis, the optimal number of clusters was determined to be 5. The Elbow plot showed an inflection point at 4, suggesting a diminishing results after this point. However, the silhouette score peaked at k=5, suggesting this configuration yields the most wel-defined and cohesive clusters. Hence, k=5 was selected, which balances model interpretability with segmentation performance.

```python

k_final = 5
kmeans = KMeans(n_clusters=k_final, random_state=42, n_init=10)
customer_df['Cluster_KMeans'] = kmeans.fit_predict(X_scaled)
```

```python

customer_df[['Customer ID', 'Cluster_KMeans'] + features].head()
```

```python

for feature in features:
    plt.figure(figsize=(7, 4))
    sns.boxplot(data=customer_df, x='Cluster_KMeans', y=feature, hue='Cluster_KMeans', palette='Set2', legend=False)
    plt.title(f'{feature} by Cluster')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

The cluster analysis revealed clear differences in customer behaviour. Cluster 4 consistently demonstrated high engagement, recent activity and the highest CLV, thereby making it the most profitable segment. In contrast, Cluster 2 had the highest recency and the low frequency, suggesting a possibility of churning. Age and unit cost patterns suggest further segmentation opportunities based on demographics and preferences in products. These insights can further guide marketing strategies to amplify retention, re-engagement, and high-value targeting.

```python
# Full hierarchical clustering on the entire dataset (approx. 950,000 samples) is memory intensive,
# and causes runtime crashes on limited-resource environments.
# To avoid this, I sampled 1,000 customers and applied linkage and dendrogram generation, keeping it interpretable and representable.

hc_sample = customer_df.sample(n=1000, random_state=42)
hc_scaled = scaler.transform(hc_sample[features])
hc_model = AgglomerativeClustering(n_clusters=5)
hc_sample['Cluster_HC'] = hc_model.fit_predict(hc_scaled)

linked = linkage(hc_scaled, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='level', p=20,
           leaf_rotation=90, leaf_font_size=10)
plt.title('Hierarchical Clustering Dendrogram (Sample of 1000)')
plt.xlabel('Sample index or Cluster Merge')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()

```

The dendrogram reinforces the 4-5 cluster hypothesis observed from the Elbow and Silhouette methods. The larger verticle distances between cluster merges suggest a more natural seperation around 4 and 5 groups, validating the consistency across the clustering techniques

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
customer_df['PCA1'] = X_pca[:, 0]
customer_df['PCA2'] = X_pca[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=customer_df, x='PCA1', y='PCA2', hue='Cluster_KMeans', palette='tab10', edgecolor='white', alpha=0.7)
plt.title("KMeans Clusters (PCA 2D)")
plt.grid(True)
plt.show()

explained_variance = pca.explained_variance_ratio_.sum()
print(f"Total variance explained by the first 2 components: {explained_variance:.2%}")
```

PCA was implemented to reduce the feature space into two principal components. The 2D projection shows resonably distinct clusters, especially 0 and 4, which are more spacially distanced from the centre-region of the clusters. This suggests the PCA was effective at preserving the separability of major customer groups.

```python
#t-SNE is applied to a random sample of 5,000 customers to balance visual clarity and computation
tsne_sample = customer_df.sample(n=5000, random_state=42)
tsne_scaled = scaler.transform(tsne_sample[features])
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=1000)
X_tsne = tsne.fit_transform(tsne_scaled)

tsne_sample['TSNE1'] = X_tsne[:, 0]
tsne_sample['TSNE2'] = X_tsne[:, 1]

plt.figure(figsize=(8, 6))
sns.scatterplot(data=tsne_sample, x='TSNE1', y='TSNE2', hue='Cluster_KMeans', palette='tab10', edgecolor='white', alpha=0.7)
plt.title("KMeans Clusters (t-SNE 2D - Sample of 5000)")
plt.grid(True)
plt.show()

```

t-SNE out performed PCA for visual separation of clusters, making it a more effective tool for interpreting customer segments in this context. However, since t-SNE does not preserve global distances, PCA remains valuable for understanding overall variance structure.
